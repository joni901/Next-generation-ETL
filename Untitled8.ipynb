```python
import pandas as pd

# URL RAW do arquivo JSON
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/main/TelecomX_Data.json"

# Leitura do JSON
df = pd.read_json(url)

# Exibir as 5 primeiras linhas
print(df.head())
```

```python
# Extrair as colunas que s√£o dicion√°rios e normalizar
customer_df = pd.json_normalize(df['customer'])
phone_df = pd.json_normalize(df['phone'])
internet_df = pd.json_normalize(df['internet'])
account_df = pd.json_normalize(df['account'])

# Agora juntar tudo no df original, retirando as colunas antigas
df_final = pd.concat([df.drop(['customer', 'phone', 'internet', 'account'], axis=1),
                      customer_df, phone_df, internet_df, account_df], axis=1)

print(df_final.head())
```

```python
# Detectar colunas com valores NaN
colunas_com_nan = df.columns[df.isna().any()].tolist()

# Detectar colunas com strings vazias
colunas_com_vazio = df.columns[(df.astype(str).apply(lambda x: x.str.strip() == '')).any()].tolist()

print("Colunas com NaN:", colunas_com_nan)
print("Colunas com strings vazias:", colunas_com_vazio)
```

```python
# Remover linhas com valores NaN nas colunas
df = df.dropna(subset=['Churn'])

# Remover linhas onde alguma dessas colunas √© string vazia
df = df[~df[['Churn']].astype(str).apply(lambda x: x.str.strip() == '').any(axis=1)]

```

```python
# Detectar colunas com valores NaN
colunas_com_nan = df.columns[df.isna().any()].tolist()

# Detectar colunas com strings vazias
colunas_com_vazio = df.columns[(df.astype(str).apply(lambda x: x.str.strip() == '')).any()].tolist()

print("Colunas com NaN:", colunas_com_nan)
print("Colunas com strings vazias:", colunas_com_vazio)
```

```python

# Excluir linhas onde 'Churn' est√° ausente (nulo)
df = df.dropna(subset=['Churn'])

# Verificar se ainda h√° valores ausentes
print("\nüîç Valores ausentes ap√≥s a limpeza:")
print(df.isnull().sum())

# (Opcional) Salvar o DataFrame limpo
df.to_csv("TelecomX_Data_limpo.csv", index=False)
print("\n‚úÖ Linhas com Churn nulo foram removidas e novo arquivo salvo.")


```
